# K-Means

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample_df = pd.read_csv('r15.csv')
sample_df.head()
```

```python
data = sample_df[['col1','col2']]
target = sample_df['target']
```

```python
# k의 개수가 15개
km = KMeans(n_clusters=15).fit(data)
```

```python
plt.scatter(data['col1'], data['col2'], c=km.labels_,
           cmap='rainbow')
plt.show()
```

### 다른 데이터 예시

```python
sample_df2 = pd.read_csv('spiral.csv')
sample_df2.head()
```

```python
data2 = sample_df2[['col1','col2']]
target2 = sample_df2['target']
```

```python
plt.scatter(data2['col1'], data2['col2'], c=km.labels_,
           cmap='rainbow')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
# 반지름이 0.6, 최소 샘플 10개
dbscan = DBSCAN(eps=0.6, min_samples=10).fit(data)
```

```python
plt.scatter(data['col1'], data['col2'], c=dbscan.labels_,
           cmap='rainbow')
plt.show()
```

##### K-Means에 비해 중간 데이터 분별이 안되고 있습니다. 파라미터 조절이 필요해 보입니다.

```python
# 반지름이 0.6, 최소 샘플 10개
dbscan = DBSCAN(eps=2, min_samples=3).fit(data2)
```

```python
plt.scatter(data2['col1'], data2['col2'], c=dbscan.labels_,
           cmap='rainbow')
plt.show()
```

##### 이 데이터셋은 K-Means 보다 DBSCAN에서 분류가 더 정확해 보입니다. 각 데이터셋이 맞는 기술을 사용하면 될 것 같습니다.
